```python
#급여가 2000보다 큰 데이터만 출력
result = db.emp.find(
    {
        "SAL": {
            "$gt":2000
        }
    }
)
for list in result:
    print(list)
```

```python
#급여가 2450보다 작은 데이터만 출력
result = db.emp.find(
    {
        "SAL": {
            "$lte":2450
        }
    }
)
for list in result:
    print(list)
```

```python
#SELECT * FROM emp WHERE ENAME = SMITH OR ENAME = ALLEN

result = db.emp.find(
    {'$or':
    [{'ENAME':'SMITH', "ENAME":'ALLEN'}]})
for list in result:
    print(list)
```

```python
# SELECT * FROM emp WHERE ENAME=SMITH OR ENAME=ALLEN
result = db.emp.find(
    {
        "$or":[
            {
                "ENAME" : "SMITH"
            },
            {
                "ENAME" : "ALLEN"
            }
        ]
    }
)
for list in result:
    print(list)
```

```python
#SELCT * FROM emp WHERE ENAME = SMITH AND JOB = CLERK
result = db.emp.find(
    {
                "ENAME" : "SMITH",
                "JOB" : "CLERK"
    }
)
for list in result:
    print(list)
```

```python
# SELECT * FROM emp WHERE SAL >= 2000 AND SAL < 4000
result = db.emp.find(
    {
        "SAL" : {
            "$gte" : 2000,
            "$lt" : 4000
        }
    }, {
        "_id" : 0
    }
)

pd.DataFrame(result)
```

```python
#SELECT * FROM emp WHERE ENAME LIKE %S%
result = db.emp.find(
    {
        "ENAME" : {
            "$regex" : "S"
        }
    }
)
pd.DataFrame(result)
```

```python
#SELECT * FROM emp WHERE ENAME LIKE %S
result = db.emp.find(
    {
        "ENAME" : {
            "$regex" : "^S"
        }
    }
)
pd.DataFrame(result)
```

```python
#SELECT * FROM emp WHERE ENAME LIKE %S%
result = db.emp.find(
    {
        "ENAME" : {
            "$regex" : "S$"
        }
    }
)
pd.DataFrame(result)
```

```python
'''
SELECT ENAME, JOB, SAL, DEPTNO FROM emp
WHERE SAL > 1250 AND 
SAl <= 3000 AND
JOB = SALESMAN
ORDER BY ENAME
'''
result = db.emp.find(
    {
        "SAL" : {
            "$gt" : 1250,
            "$lte" : 3000
        },
        "JOB": "SALESMAN"
    },
    {
        "ENAME" : 1,
        "JOB" : 1,
        "SAL" : 1,
        "DEPTNO" : 1,
        "_id" : 0
    }
).sort("ENAME", -1)

pd.DataFrame(result)
```
