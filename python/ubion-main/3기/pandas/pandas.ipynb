```python
# 코로나 누적확진률이 3.5 이상이면 High, 3.5 미만이면 Low 컬럼을 추가

corona_df_5 = corona_df_3.copy()
corona_df_5["H/L"] = corona_df_5["누적확진률"].apply(lambda x: "High" 
                    if (x >= 3.5) else "Low")
corona_df_5
```

```python
def change(x):
    if x <3:
        return 0
    elif x < 5:
        return 1
    elif x < 7:
        return 2
    else:
        return 3

test.A = test.A.apply(change)
test
```

```python
corona_df_5["등록일시"] = pd.to_datetime(corona_df_5["등록일시"], 
                                        format="%Y-%m-%d")


corona_df_5["요일"] = corona_df_5["등록일시"].dt.weekday
corona_df_5
```

```python
def week_change(x):
    if x == 0:
        return "월"
    elif x ==1:
        return "화"
    elif x ==2:
        return "수"
    elif x ==3:
        return "목"
    elif x ==4:
        return "금"
    elif x ==5:
        return "토"
    elif x ==6:
        return "일"
    else:
        return

corona_df_5["요일"] = corona_df_5["요일"].apply(week_change)
```
